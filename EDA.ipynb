```python
#5. Análisis de correlaciones
#Permite identificar relaciones entre variables (por ejemplo, si un aumento en la RoomRate se relaciona con un aumento en Revenue). Las correlaciones cercanas a 1 o -1 indican relaciones fuertes.

# Matriz de correlación
correlation_matrix = df[['RoomRate', 'CleaningFee', 'Revenue', 'NetRevenue', 'NumNights', 'Capacity']].corr()

# Mapa de calor para visualizar la correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación entre Variables Numéricas')
plt.show()
```

```python
# Función para detectar outliers usando el IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Detección de outliers en ingresos y duración de estadía
outliers_revenue = detect_outliers(df, 'NetRevenue')
outliers_nights = detect_outliers(df, 'NumNights')

print(f"Outliers en ingresos netos: {len(outliers_revenue)}")
print(f"Outliers en duración de estadía: {len(outliers_nights)}")
```
