```python
#CAMBIO EL FORMATO DE LA FECHA DE ReadyDate Y A LAS FILAS VACIAS DE 'PropertyType' LAS LLENO CON LA PALABRA 'Unknown'

df_airbnb_us['ReadyDate'] = pd.to_datetime(df_airbnb_us['ReadyDate'])
df_airbnb_us['PropertyType'] = df_airbnb_us['PropertyType'].fillna('Unknown')
# Reemplazar 'Apa' por 'Apartment' en la columna 'PropertyType'
df_airbnb_us['PropertyType'] = df_airbnb_us['PropertyType'].replace('Apa', 'Apartment')
df_airbnb_us.info()
```

```python
#CAMBIO EL FORMATO DE LAS COLUMNAS QUE TIENEN FECHAS

# Función para convertir fechas
def convert_date(date_str):
    """
    Convierte una cadena de fecha en un objeto datetime.

    Intenta convertir la fecha en el formato 'YYYY-MM-DD HH:MM:SS'.
    Si falla, intenta convertirla en el formato 'DD/MM/YYYY'.
    
    Args:
        date_str (str): Cadena que representa una fecha.

    Returns:
        pd.Timestamp: Objeto datetime correspondiente a la fecha,
                       o NaT si la conversión falla.
    """
    try:
        # Intenta convertir la fecha en el formato completo
        return pd.to_datetime(date_str, format='%Y-%m-%d %H:%M:%S')
    except ValueError:
        # Si falla, intenta convertirla en el formato corto
        return pd.to_datetime(date_str, format='%d/%m/%Y', errors='coerce')

# Aplicar la función a la columna de fechas
df_booking_us['BookingCreatedDate'] = df_booking_us['BookingCreatedDate'].apply(convert_date)

# Verificar el resultado
df_booking_us.info()
```

```python
df_booking_us.head()
```

```python
#RELLENO LOS LUGARES VACIOS CON VALORES ACORDES A LOS NECESARIOS

df_booking_us['ArrivalDate'] = pd.to_datetime(df_booking_us['ArrivalDate'])
df_booking_us['DepartureDate'] = pd.to_datetime(df_booking_us['DepartureDate'])
df_booking_us['RoomRate'] = df_booking_us['RoomRate'].fillna(df_booking_us['TotalPaid']/df_booking_us['NumNights'])
df_booking_us['Revenue'] = df_booking_us['Revenue'].fillna(df_booking_us['TotalPaid'] - df_booking_us['CleaningFee'])
df_booking_us['Channel'] = df_booking_us['Channel'].fillna('Unknown')
df_booking_us.info()
```

```python
# MANEJO VALORES NULOS DE ADR
# Imputar el ADR (Average Daily Rate) como Revenue/NumNights si no está disponible
df_booking_us['ADR'] = df_booking_us.apply(lambda row: row['Revenue'] / row['NumNights'] if pd.isna(row['ADR']) and row['NumNights'] > 0 else row['ADR'], axis=1)

# Imputar TouristTax con la media o una regla basada en TotalPaid si es necesario
df_booking_us['TouristTax'] = df_booking_us['TouristTax'].fillna(df_booking_us['TotalPaid'] * 0.05)  # Suponemos una tasa turística del 5%

# Generar nuevas métricas
# Ingreso neto (Revenue - CleaningFee - TouristTax)
df_booking_us['NetRevenue'] = df_booking_us['Revenue'] - df_booking_us['CleaningFee'] - df_booking_us['TouristTax']
```

```python
# Supongamos que df_prueba ya está definido y contiene tus datos

# Limpiar la columna Precio
df_prueba['Precio'] = df_prueba['Precio'].replace(r'Total sin incluir impuestos\n\$(\d{1,3}(?:,\d{3})*) COP', r'\1', regex=True)  # Extraer el precio
df_prueba['Precio'] = df_prueba['Precio'].replace(',', '', regex=True)  # Eliminar comas
df_prueba['Precio'] = pd.to_numeric(df_prueba['Precio'], errors='coerce')  # Convertir a numérico

# Limpiar la columna Huéspedes
df_prueba['Huéspedes'] = df_prueba['Huéspedes'].replace(r'.*?(\d+).*', r'\1', regex=True)  # Extraer solo el número
df_prueba['Huéspedes'] = pd.to_numeric(df_prueba['Huéspedes'], errors='coerce')  # Convertir a numérico

# Limpiar la columna Habitaciones
df_prueba['Habitaciones'] = df_prueba['Habitaciones'].replace(r'.*?(\d+).*', r'\1', regex=True)  # Extraer solo el número
df_prueba['Habitaciones'] = pd.to_numeric(df_prueba['Habitaciones'], errors='coerce')  # Convertir a numérico

# Limpiar la columna Camas
df_prueba['Camas'] = df_prueba['Camas'].replace(r'.*?(\d+).*', r'\1', regex=True)  # Extraer solo el número
df_prueba['Camas'] = pd.to_numeric(df_prueba['Camas'], errors='coerce')  # Convertir a numérico

# Limpiar la columna Baños
df_prueba['Baños'] = df_prueba['Baños'].replace(r'.*?(\d+).*', r'\1', regex=True)  # Extraer solo el número
df_prueba['Baños'] = pd.to_numeric(df_prueba['Baños'], errors='coerce')  # Convertir a numérico

# Limpiar la columna Puntuación
df_prueba['Puntuación'] = df_prueba['Puntuación'].str.extract(r'(\d+\.\d+|\d+)')  # Extraer números, incluyendo decimales
df_prueba['Puntuación'] = pd.to_numeric(df_prueba['Puntuación'], errors='coerce')  # Convertir a numérico

# Revisión de valores nulos
print(df_prueba.isnull().sum())  # Muestra el conteo de valores nulos por columna
```

```python
# RELLENO LOS NULOS QUE NO TIENEN ZONAS QUE COINCIDAN CON EL PROMEDIO TOTAL DE LA PUNTUACION (4.59) ADEMAS CON EL PROMEDIO DE BA;OS Y DE PRECIOS


# Calcular la media de la columna 'Puntuación'
puntuacion_promedio1 = df_prueba['Puntuación'].mean()
baños_promedio = df_prueba['Baños'].mean()
precio_promedio = df_prueba['Precio'].mean()
# Rellenar los NaN en la columna 'Puntuación' con la media
df_prueba['Puntuación'] = df_prueba['Puntuación'].fillna(puntuacion_promedio1)
df_prueba['Baños'] = df_prueba['Baños'].fillna(baños_promedio)
df_prueba['Precio'] = df_prueba['Precio'].fillna(precio_promedio)

# Redondear los valores de 'Puntuación' a 2 decimales
df_prueba['Puntuación'] = df_prueba['Puntuación'].round(2)


# Mostrar los primeros 68 registros
df_prueba.head(68)
```
